<h2 style = 'color:#white;font-size:20px'>4. Видаляємо елемент з дата фрейму </h2>

```python
df.drop(columns = 'index',inplace=True)
```

<h2 style = 'color:#white;font-size:20px'>5. Які авіакомпанії вказано в наборі даних, а також їхні частоти?</h2>

```python
df_group = df['airline'].value_counts()

df_group
```

<h2 style = 'color:#white;font-size:20px'>6. Чи залежить ціна від авіакомпанії?</h2>

```python
df_price = df.groupby('airline')['price'].agg(['max','min','mean','std','median'])

df_price
```

<p style="font-size:14px">Виконавши код вище чітко можна побачити що є дві компанії які мають вищу ціну як найнищу так і середеню, так і медіанну (що говорить про те що викиди не сильно впливають на результати вибірки)<p>

<h2 style = 'color:#white;font-size:20px'>7. Чи змінюється ціна квитка залежно від часу відправлення та часу прибуття?</h2>

```python
df_price = df.pivot_table(values='price',columns='departure_time',index='arrival_time',aggfunc=["mean"])

df_price
```

<p style="font-size:14px;margin-top:20px">В результатх видно що найменше коштують квитки які прибувають у пізній час або відправляються, так само і рейсами які йдуть надто рано<p>




<h2 style = 'color:#white;font-size:20px'>8. Як впливає на ціну купівля квитків лише за 1 або 2 дні до відправлення?</h2>

```python
df_price = df.groupby('days_left')['price'].agg(['max','min','mean','std','median'])

df_price
```

<p style='font-size:14px'>Досить цікава статистика яка говорить що купувати квиток за день до рейсу вигідніше чим за 2 тижні</p>

<p style='font-size:14px'>Але ця статистика не заперечує і того що купувати пізніше означає дешевше, чим далі від початку рейсу тим дешевші квитки</p>

<h2 style = 'color:#white;font-size:20px'>9. Як відрізняється ціна квитка між економ-класом та бізнес-класом?</h2>

```python
df_price = df.groupby('class')['price'].agg(['max','min','mean','std','median'])

df_price
```

<p style='font-size:14px'>Чітко видно різницю що бізнес-клас набагато дорожчий чим економ, можна сказати приблизно в 10 разів, але є і економ клас який коштує як бізнес</p>
