<h2 style = 'color:#white;font-size:20px'>1. Імпорт даних та читання <b><i>csv</i></b> файлу </h2>

```python
import pandas as pd
df = pd.read_csv('C:/Users/den/Desktop/pandas/pandas-learn/data/Home_Sale_Data.csv')

df.info()
```

<h2 style = 'color:#white;font-size:20px'>2. Cтворення стовбця комісії як 10% від стовпця <b><i>SalePrice</i></b> </h2>

```python
df['Commision'] = df['SalePrice'] * 0.1

df[['Address','Commision']]
```

<h2 style = 'color:#white;font-size:20px'>3. Cтворення стовбця різнці цін між запрошуваною ціною та ціною продажу</h2>

```python
df['Difference'] = df['AskingPrice'] - df['SalePrice']

df[['AskingPrice','SalePrice','Difference']]
```

<h2 style = 'color:#white;font-size:20px'>4. Cтворення стовбця відсоткової різниці між стовпцем різниці та запрошуваною ціною</h2>

```python
df['Margin'] = df['Difference']/df['AskingPrice'] *100

df[['Margin','Difference','AskingPrice']]
```

<h2 style = 'color:#white;font-size:20px'>5. Заокруглення за допомогою методу <b><i>round</i></b> до 3 знаків</h2>


```python
df['Margin'] = df['Margin'].round(3) 

df['Margin']
```

<h2 style = 'color:#white;font-size:20px'>6. Визначення умовного стовпця за допомогою комісії</h2>

```python

def deal_type_fun(el):
    if el>50000:
        return 'High Deal'
    else:
        return 'Regular Deal'

df['DealType'] = df['Commision'].apply(deal_type_fun)

df[['Commision','DealType']]
```
